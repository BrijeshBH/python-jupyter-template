### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11, 1)
# Define y values as the square of x
y = x ** 2

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customize the line plot
plt.title('Square Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values (X Squared)')
plt.grid(True)
plt.legend()

# Show the line plot
plt.show()

# Define bar plot data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the bar plot
plt.title('Category Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the bar plot
plt.show()

# Define scatter plot data
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x_scatter, y_scatter, color='blue')

# Annotate each point
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(f'{txt}', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Customize the scatter plot
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

# Show the scatter plot
plt.show()

# Define pie chart data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'lightgreen', 'orange'])

# Customize the pie chart
plt.title('Programming Language Popularity')

# Show the pie chart
plt.show()
```
