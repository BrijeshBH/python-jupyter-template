### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Define x values from 0 to 10
x = np.arange(0, 11, 1)
# Define y values as the square of x
y = x ** 2

# Create the line plot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')
axs[0, 0].set_title('Square Function Plot')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values (X Squared)')
axs[0, 0].grid(True)
axs[0, 0].legend()

# Define bar plot data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Category Values')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Define scatter plot data
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Create the scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    axs[1, 0].annotate(f'{txt}', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].grid(True)

# Define pie chart data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'lightgreen', 'orange'])
axs[1, 1].set_title('Programming Language Popularity')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
