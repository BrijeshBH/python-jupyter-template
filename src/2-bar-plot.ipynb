### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11, 1)
# Define y values as the square of x
y = x ** 2

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customize the line plot
plt.title('Square Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values (X Squared)')
plt.grid(True)
plt.legend()

# Show the line plot
plt.show()

# Define bar plot data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the bar plot
plt.title('Category Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the bar plot
plt.show()
```
